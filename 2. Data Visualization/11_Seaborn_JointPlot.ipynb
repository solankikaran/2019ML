```python
import seaborn as sns #importing as sns is convention
%matplotlib inline #Used for visualization inside the %notebook
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

jointplot() is used for bivariate data (Two variables)

```python
sns.jointplot(x='total_bill', y='tip', data=tips) #Takes three parameters.
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') #kind by default has scatter. we'll put it as hex.
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') #Draws a regression line.
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde') #Shows density where points match up the most
```
