```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.plot(x2, y2, 'o-')
plt.title('Subplot 1')
plt.xlabel('X1')
plt.ylabel('Amplitude')
plt.show
```
