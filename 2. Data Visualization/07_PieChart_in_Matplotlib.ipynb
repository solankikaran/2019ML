```python
import matplotlib.pyplot as plt
```

```python
activities = ['Eat', 'Sleep', 'Work', 'Play']
```

```python
slices = [3, 4, 8, 6]
color = ['g', 'r', 'b', 'y']
```

```python
plt.pie(slices, labels=activities, colors=color, startangle=90, shadow=True, explode=(0, 0, 0, 0.1),autopct='%1.2f%%')

#startangle starts from 90 degree
#shadow=True gives shadow to the Pie Chart
#explode takes in a tuple and removes/explodes a slice from Pie Chart
#autopct function shows in percentange 1.2f means 2 values are shown after decimal point

plt.legend()
plt.show()
```
