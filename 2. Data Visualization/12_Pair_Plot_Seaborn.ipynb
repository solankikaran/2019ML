```python
import seaborn as sns #importing as sns is convention
%matplotlib inline #Used for visualization inside the %notebook
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex') #Hue takes in a column with categorical data
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') #palette is for different colors.
```
