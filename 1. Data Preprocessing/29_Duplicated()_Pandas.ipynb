```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df['Senior Management'] = df['Senior Management'].astype(bool)
df.sort_values('First Name', inplace=True)
df['Gender'] = df['Gender'].astype("category")
```

```python
df['First Name'].duplicated() #Considers the first occurence as False(Not Duplicate). and the ones after it as True (duplicate)
```

```python
df['First Name'].duplicated(keep='last') #Will consider the first occurence from bottom. then as duplicate(True)
```

```python
~df['First Name'].duplicated(keep='last') #Tilde ~ does the reverse. Also keep=False is also correct
```
