```python
import pandas as pd
```

```python
nba = pd.read_csv('nba.csv').dropna(how='all')
```

```python
nba['Salary'] = nba['Salary'].fillna(0).astype(int) #astype will fill 0 as int instead of float
```

```python
nba.head()
```

```python
nba['Salary'].rank(ascending=False).astype(int) #Ranks playerd according to salary
```

```python
nba['Salary_Rank'] = nba['Salary'].rank(ascending=False).astype(int)
```
