```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv') #Use df henceforth.
```

```python
df.info()
```

```python
df['Start Date'] #Here Data type is wrongly given as Object
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date']) #Converts dtype to datetime which is correct
```

```python
df.info() #You can see Start Date is converted to datetime!
```

```python
df['Last Login Time'].head() #Here dtype is given as object. 
```

## Note: Strings are considered as objects. Last Login time is by no means a String

```python
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
```

```python
df.info()
```

```python
df['Last Login Time']
```

### Senior Management is boolean. We'll change that now.

```python
df['Senior Management'].head()
```

```python
df['Senior Management'] = df['Senior Management'].astype(bool) #Converted to bool
```

```python
df.info()
```

### Gender is also given as an object. But its a category.

```python
df['Gender'].head()
```

```python
df['Gender'] = df['Gender'].astype("category")
```

```python
df.info()
```

### Q: Now we had to code all these things. Why?
Ans: Look at the df.info() before we did the operations. Look at memory usage. See the difference?
It was 62+ KB with improper dtypes, but after properly assigning dtypes, memory has been reduced to 49 Kb+.
Hence filtering improves the memory management! Therefore its important that you filter data according to its data type!

### Pst: There's actually a short way of doing this instead of all this donkey-work. 

How? This Way - >

```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df['Senior Management'] = df['Senior Management'].astype(bool)
df['Gender'] = df['Gender'].astype("category")
```

```python
df.info()
```
