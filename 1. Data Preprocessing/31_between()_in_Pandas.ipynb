```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df['Senior Management'] = df['Senior Management'].astype(bool)
df.sort_values('First Name', inplace=True)
df['Gender'] = df['Gender'].astype("category")
```

```python
df.head()
```

```python
df['Salary'].between(60000, 70000) #Gives people having salary between 60k-70k
```

```python
df[df['Salary'].between(60000, 70000)]
```

```python
df[df['Bonus %'].between(1.5, 4.5)] #Gives people having bonus % between 1.5 and 4.5
```

### .between() also works with datetime also!
